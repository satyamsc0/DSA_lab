### Implement Quick sort

```python
# Function to find the partition position 
def partition(array, low, high): 
    # Choose the rightmost element as pivot 
    pivot = array[high] 

    # Pointer for greater element 
    i = low - 1 
    # Traverse through all elements and compare each element with pivot 
    for j in range(low, high): 
        if array[j] <= pivot: 
            # If element smaller than pivot is found 
            # swap it with the greater element pointed by i 
            i = i + 1 
            # Swapping element at i with element at j 
            (array[i], array[j]) = (array[j], array[i]) 

    # Swap the pivot element with the greater element specified by i 
    (array[i + 1], array[high]) = (array[high], array[i + 1]) 

    # Return the position from where partition is done 
    return i + 1 

# Function to perform quicksort 

def quicksort(array, low, high): 
    if low < high: 
        # Find pivot element such that element smaller than pivot are on the left and element greater than pivot are on the right 
        pi = partition(array, low, high) 
        
        # Recursive call on the left of pivot 
        quicksort(array, low, pi - 1) 

        # Recursive call on the right of pivot 
        quicksort(array, pi + 1, high) 

array = [10, 60, 40, 50, 20, 30] 
print("Unsorted array: ",array)
N = len(array) 

# Function call 
quicksort(array, 0, N - 1) 

print("Sorted array: ",array) 
```

#### Implement Merge sort

```python
array = [15,4,3,2,6] 

print("Unsorted array: ", array) 

# Divide the array into halfs at every step untill it is not divisible 

def mergeSort(array): 

    if len(array)>1: 

        mid = len(array)//2 

        L = array[:mid] 
        R = array[mid:] 
        
        mergeSort(L) 
        mergeSort(R) 

        # i pointer for L 
        #j pointer for R 
        #k pointer for main array 

        i=j=k=0 

        # take the min element from the subarray and merge it in the main array 

        while i<len(L) and j<len(R): 

            if L[i]<=R[j]: 
                array[k]=L[i] 
                i+=1 
            else: 
                array[k]=R[j] 
                j+=1 
            k+=1 

        # insert the remaining values of L (if any) in main array     
        while i<len(L): 
            array[k] = L[i] 
            k+=1 
            i+=1 

        # insert the remaining values of R (if any) in main array 

        while j<len(R): 
            array[k]=R[j] 
            j+=1 
            k+=1 

mergeSort(array) 
print("Sorted array: ", array) 
```

#### Implementation of the “Stack using an array”. Use these concept: Push(value), Pop(), FULL, EMPTY, and FindTOP position.

```python
# Push operation
def Push(stack, value, stack_size):
    if len(stack) < stack_size:
        stack.append(value)
    else:
        print("Stack Overflow") 
    return stack

# Pop operation
def Pop(stack):
    if len(stack)==0:
        print("Stack Underflow")
        return 
    else:
        poped_ele = stack[-1]
        print("Poped element: ",poped_ele)
        del stack[-1]
        return stack

# Full operation
def isFull(stack,stack_size):
    if len(stack)==stack_size:
        print("The stack is full")
    else:
        print("The stack is not full")

# Empty operation
def isEmpty(stack):
    if len(stack)==0:
        print("The stack is empty")
    else:
        print("The stack is not empty")

# findTop operation
def findTop(stack):
    if len(stack)==0:
        print("The stack is empty")
    else:
        print("The top element is: ",stack[-1])
        

# Defining the stack size
stack_size = int(input("Enter the size of the stack: "))
stack = []

# A custom loop just to perform the above operations based on user's input
while True:
    user_input = int(input("""Enter 1 to push an element in the stack
Enter 2 to pop the element from the stack
Enter 3 to check whether the stack is full or not
Enter 4 to check whether the stack is empty or not
Enter 5  to find the top element
Enter any number to exit\n"""))
    if user_input == 1:
        value = int(input("enter the value: "))
        stack = Push(stack, value, stack_size)
        print("Stack: ",stack)
        
    elif user_input == 2:
        stack = Pop(stack)
        print("Stack: ",stack)
        
    elif user_input == 3:
        isFull(stack,stack_size)
        
    elif user_input == 4:
        isEmpty(stack)
        
    elif user_input == 5:
        findTop(stack)
    else:
        break
```
