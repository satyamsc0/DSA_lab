#### Implementation of  “Single Linked List”.

```python
# Class to create a node
class Node: 
    def __init__(self, data): 
        self.data = data 
        self.next = None 

# Linked List class 
class LinkedList: 
    
    def __init__(self): 
        self.head = None 

    # Function to insert a new node at the beginning 
    def insertBegin(self, new_node): 
        # If LL is empty
        if self.head is None: 
            self.head = new_node 
        else: 
            current = self.head 
            new_node.next = current 
            self.head = new_node                     

    # Function to insert a new node at the end 
    def insertEnd(self, new_node): 
        if self.head is None: 
            self.head = new_node 
        else: 
            current = self.head 
            # Iterating till the last node
            while current.next is not None: 
                current = current.next 
            
            current.next = new_node 
            
    # Function to insert a new node at any position 
    def insertAtAnyPos(self, pos, new_node): 
        # If Ll is empty and user wants to insert at the 1st pos
        if self.head is None and pos==1: 
            new_node.next = self.head 
            self.head = new_node 
        else: 
            # if LL is not empty
            flag = False 
            current = self.head 
            prev = current 
            current_pos = 1 
            while current is not None:  # Till node exists               
                if pos == 1: 
                    # if its 1st pos
                    self.insertBegin(new_node) 
                    flag = True 
                    break 
                    # if pos is somewhere in between
                elif current_pos == pos: 
                    prev.next = new_node 
                    new_node.next = current 
                    flag = True 
                    break 
                else: 
                    # iterating
                    current_pos +=1 
                    prev = current 
                    current = current.next 
            
            if flag == True:
                # insertion successful
                pass
            elif flag == False and current_pos == pos: 
                # if user wants to insert at the last pos
                prev.next = new_node 
            else: 
                print("Invalid position") 

    # Function to delete a node at the beginning 
    def deleteBegin(self): 
        if self.head is None: 
            print("Can't delete, Linked list is empty") 
        else: 
            # moving the head pointer to the 2nd node
            self.head = self.head.next 
    
    # Function to delete node at the end 
    def deleteEnd(self): 
        if self.head is None: 
            print("Can't delete, Linked list is empty") 
        else: 
            current = self.head 
            prev = current 
            while current.next is not None: 
                prev = current 
                current = current.next 
            
            # if only 1 node exists
            if self.head.next is None: 
                self.head = None 
            else: 
                prev.next = None 

    # Function to delete a node at any position 
    def deleteAtAnyPos(self, pos): 
        if self.head is None: 
            print("Can't delete, Linked list is empty") 
        else: 
            flag = False 
            current = self.head 
            prev = current 
            current_pos = 1 
            while current is not None: 
                if pos == 1: 
                    self.deleteBegin() 
                    flag = True 
                    break 
                elif current_pos == pos: 
                    prev.next = current.next 
                    current.next = None
                    flag = True 
                    break 
                else: 
                    current_pos +=1 
                    prev = current 
                    current = current.next 
                    
            if flag == False: 
                print("Invalid position") 
             
    # Function to print the linkedlist 
    def printLL(self): 
        ll = '' 
        current = self.head 
        while current: 
            data = current.data 
            ll = ll + f'{data} -->' 
            current = current.next 
        ll += 'NULL' 
        print("Linked List: ",ll) 

# Code execution starts here 
if __name__=='__main__': 
    # Start with the empty list 
    ll = LinkedList() 
    while True: 
        user_input = int(input("""Enter 1 to insert at beginning 
Enter 2 to insert at end 
Enter 3 to insert at any position 
Enter 4 to delete at beginning 
Enter 5 to delete at end 
Enter 6 to delete at any position 
Enter 7 to display the Linked list 
Enter any number to exit\n""")) 
        if user_input == 1: 
            data = int(input("Enter the node data: ")) 
            new_node = Node(data) 
            ll.insertBegin(new_node) 
        elif user_input == 2: 
            data = int(input("Enter the node data: ")) 
            new_node = Node(data) 
            ll.insertEnd(new_node) 
        elif user_input == 3: 
            data = int(input("Enter the node data: ")) 
            new_node = Node(data) 
            pos = int(input("Enter the position: ")) 
            ll.insertAtAnyPos(pos, new_node) 
        elif user_input == 4: 
            ll.deleteBegin() 
        elif user_input == 5: 
            ll.deleteEnd() 
        elif user_input == 6: 
            pos = int(input("Enter the position: ")) 
            ll.deleteAtAnyPos(pos) 
        elif user_input == 7: 
            ll.printLL() 
        else: 
            break 
```

#### Implementation of the "Doubly Linked List".

```python
class Node: 
    def __init__(self, data): 
        self.data = data 
        self.prev = None 
        self.next = None 

class DoublyLinkedList(): 
    def __init__(self): 
        self.head = None 

    # Function to insert a new node at the beginning 
    def insertBegin(self, new_node): 
        # if ll is empty 
        if self.head is None: 
            self.head = new_node 
        else: 
            # if there exists nodes 
            new_node.next = self.head 
            self.head.prev = new_node 
            self.head = new_node 
            
    def insertEnd(self, new_node): 
        if self.head is None: 
            self.head = new_node 
        else: 
            current = self.head 
            while current.next is not None: 
                current = current.next
                
            current.next = new_node 
            new_node.prev = current 

    def insertAtAnyPos(self, new_node, pos): 
        # if ll is empty and inserting at pos 1
        if self.head is None and pos == 1: 
            self.head = new_node 
        else: 
            # if nodes are present
            flag = False 
            current_pos = 1 
            current = self.head 
            previous = current 
            while current is not None: 
                if pos == 1: 
                    self.insertBegin(new_node) 
                    flag = True 
                    break 
                elif current_pos == pos: 
                    previous.next = new_node 
                    new_node.prev = previous 
                    new_node.next = current
                    current.prev = new_node
                    flag = True 
                    break 
                else: 
                    current_pos+=1 
                    previous = current 
                    current = current.next 
             
            # if user wants to insert at the last pos 
            if flag == False and current_pos == pos:             
                previous.next = new_node 
                new_node.prev = previous 
                # insertion successful
            elif flag == True: 
                pass 
            else: 
                print("Invalid position") 

    def deleteBegin(self): 
        if self.head is None: 
            print("Can't delete, Linked list is empty") 
        else: 
            current = self.head 
            self.head = self.head.next 
            if current.next is not None: 
                # removing the links of the deleted node
                current.next.prev = None 
                current.next = None 
 

    def deleteEnd(self): 
        if self.head is None: 
            print("Can't delete, Linked list is empty") 
        else: 
            current = self.head 
            previous = current 
            while current.next is not None: 
                previous = current 
                current = current.next 

            if self.head.next is None: 
                self.head = None 
            else: 
                previous.next = None 
                current.prev = None 
     
    def deleteAtAnyPos(self, pos): 
        if self.head is None: 
            print("Can't delete, Linked list is empty") 
        else: 
            flag = False 
            current_pos = 1 
            current = self.head 
            previous = current 
            
            if pos == 1: 
                self.deleteBegin() 
                flag = True
                
            while current.next is not None: 
                if current_pos == pos: 
                    current.prev = None 
                    previous.next = current.next 
                    current.next.prev = previous 
                    current.next = None 
                    current.prev = None
                    flag = True 
                    break 
                else: 
                    current_pos += 1 
                    previous = current 
                    current = current.next 

            if flag == True: 
                pass 
            elif flag == False and current_pos == pos: 
                current.prev = None 
                previous.next = None 
            else: 
                print("Invalid position") 

    def printLL(self): 
        temp = self.head 
        ll='' 
        while temp is not None: 
            ll +=f'{temp.data}' + '<->' 
            temp= temp.next 
        ll+='NULL' 
        print("Linked List: ", ll) 
                           
# Code execution starts here 
if __name__=='__main__': 

    # Start with the empty list 
    dll = DoublyLinkedList()      
    new_node=Node(30) 
    dll.insertBegin(new_node) 
    new_node=Node(20) 
    dll.insertBegin(new_node) 
    new_node=Node(40) 
    dll.insertEnd(new_node)
#     new_node=Node(50)
#     dll.insertAtAnyPos(new_node,4) 
#     dll.deleteBegin() 
#     dll.deleteEnd() 
    dll.deleteAtAnyPos(4) 
#     dll.deleteAtAnyPos(1) 
    dll.printLL()    
```

#### Implementation of the binary tree and perform inorder, preorder and postorder traversal

```python
class Node: 

    def __init__(self,data): 
        self.data = data 
        self.left = None 
        self.right = None 

def print_level_order(root): 
    if not root: 
        return 
    queue = [root] 
    while queue: 
        current_level = [] 
        next_level = [] 
        for node in queue: 
            current_level.append(node.data) 
            if node.left: 
                next_level.append(node.left) 
            if node.right: 
                next_level.append(node.right) 
        print(" ".join(str(val) for val in current_level)) 
        queue = next_level 

def printInorder(root): 
    if root: 
        # First recur on left child 
        printInorder(root.left) 
        # Then print the data of node 
        print(root.data, end=" ") 
        # Now recur on right child 
        printInorder(root.right) 
        
def printPreorder(root): 
    if root: 
        # Then print the data of node 
        print(root.data, end=" ") 
        # First recur on left child 
        printPreorder(root.left) 
        # Now recur on right child 
        printPreorder(root.right) 

def printPostorder(root): 
    if root: 
        # First recur on left child 
        printPostorder(root.left) 
        # Now recur on right child 
        printPostorder(root.right) 
        # Then print the data of node 
        print(root.data, end=" ") 

if __name__ == '__main__': 
    # Root node 
    root = Node(1) 
    #Level 2 
    l1 = Node(2) 
    r1 = Node(3) 
    root.left = l1 
    root.right = r1 
    # Level 3 
    l21 = Node(4) 
    r21 = Node(5) 
    l22 = Node(6) 
    r22 = Node(7) 
    l1.left = l21 
    l1.right = r21 
    r1.left = l22 
    r1.right = r22 
    
    print_level_order(root) 
    print("Inorder: ") 
    printInorder(root) 
    print("\nPreorder: ") 
    printPreorder(root) 
    print("\nPostorder: ") 
    printPostorder(root) 
```
